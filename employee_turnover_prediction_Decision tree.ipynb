```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('D:/data/HR_capstone_dataset.csv',header=0)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df = df.rename(columns = {'time_spend_company' : 'emp_tenure',
                          'Work_accident' : 'work_accident',
                          'Department' : 'department',
                          'average_montly_hours' : 'average_monthly_hours'})
```

```python
df.columns
```

```python
df = pd.get_dummies(data=df, prefix=['salary', 'dept'], columns=['salary', 'department'])
```

```python
df.head(10)
```

```python
df.info()
```

                                                 DECISION TREE - CLASSIFICATION

```python
X= df.loc[:,df.columns!='left']
y= df['left']
```

```python
type(X)
```

```python
type(y)
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn import tree
clf_tree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clf_tree.fit(X_train,y_train)
```

```python
y_pred_train = clf_tree.predict(X_train)
y_pred_test = clf_tree.predict(X_test)
```

```python
y_pred_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score
```

```python
confusion_matrix(y_train,y_pred_train)
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
confusion_matrix(y_test,y_pred_test)
```

```python
y_pred_prob = clf_tree.predict_proba(X_test)[:, 1]
```

```python
y_pred_prob
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
```

```python
auc = roc_auc_score(y_test, y_pred_prob)
```

```python
plt.plot(fpr, tpr, label=f'Decision Tree (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

We have an AUC of 0.97 suggests that the model has a strong ability to differentiate between positive and negative instances, with  only a small overlap in the predicted probabilities of the two classes.

                                             Ensemble techniques

Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier()
```

```python
bag_clf = BaggingClassifier(base_estimator=clftree,n_estimators=1000,bootstrap=True,n_jobs=-1,random_state=42)
```

```python
bag_clf.fit(X_train,y_train)
```

```python
confusion_matrix(y_test, bag_clf.predict(X_test))
```

```python
accuracy_score(y_test, bag_clf.predict(X_test))
```

```python
precision_score(y_test, bag_clf.predict(X_test))
```

```python
recall_score(y_test, bag_clf.predict(X_test))
```

Random forest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators =1000, n_jobs=-1, random_state = 42)
```

```python
rf_clf.fit(X_train,y_train)
```

```python
confusion_matrix(y_test, rf_clf.predict(X_test))
```

```python
accuracy_score(y_test,rf_clf.predict(X_test))
```

Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier(n_estimators=250,random_state=42)
```

```python
params_grid = {"max_features":[4,5,6,7,8,9,10],
                               "min_samples_split":[2,3,10],}
```

```python
grid_search =GridSearchCV(rf_clf,params_grid,n_jobs=-1,cv=5,scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
confusion_matrix(y_test, cvrf_clf.predict(X_test))
```

```python
accuracy_score(y_test,cvrf_clf.predict(X_test))
```
